## Battleship!
* The sea is represented by a $10 \times 10$ grid.
* Ships are only one cell long.
* Player A's moves will be represented by a blue square and Player B's moves will be represented by a yellow square.
    * Moves will be in the form of $(x,y)$ where $x, y \in [0,9]$.
* When a ship is shot down, the cell will turn lavendar. 

```python
# PREAMBLE
import matplotlib.pyplot as plt
import numpy as np
from numpy import random
```

```python
# Initializing Grid, Move lists, and scores
grid = np.full((10,10), 1)

A_moves = []
B_moves = []

A_points = [0]
B_points = [0]
```

```python
def print_grid():
    plt.figure(figsize=(12, 12))
    plt.imshow(grid, origin='lower',extent=[0, 10, 0, 10], cmap='Set3')
    plt.grid(True,linewidth = "1.4", color='white')
    plt.xlim([0, 10])
    plt.ylim([0, 10])
    plt.xticks(np.arange(0, 11, step=1), color='white')
    plt.yticks(np.arange(0, 11, step=1), color='white')
    plt.tick_params(color='white')
    plt.show()
```

```python
def A_turn():
    print("PLAYER A (Blue):")
    x_val = input ("Enter x value: ")
    y_val = input("Enter y value: ")
    x = int(x_val)
    y = int(y_val)
    move = tuple((x,y))
    if move in A_moves:
        print("You already made this move. Try again")
        x_val = input ("Enter x value: ")
        y_val = input("Enter y value: ")
        x = int(x_val)
        y = int(y_val)
        move = tuple((x,y))
        A_moves.append(move)
    else:
        A_moves.append(move)
    if grid[x,y] != 1:
        grid[x,y] = -1
        A_points[0] += 1
    else:
        grid[x,y] = -2
      
    
def B_turn():
    print("PLAYER B (Yellow):")
    x_val = input ("Enter x value: ")
    y_val = input("Enter y value: ")
    x = int(x_val)
    y = int(y_val)
    move = tuple((x,y))
    if move in B_moves:
        print("You already made this move. Try again")
        x_val = input ("Enter x value: ")
        y_val = input("Enter y value: ")
        x = int(x_val)
        y = int(y_val)
        move = tuple((x,y))
        B_moves.append(move)
    else:
        B_moves.append(move)
    if grid[x,y] != 1:
        grid[x,y] = -1
        B_points[0] += 1
    else:
        grid[x,y] = 4
```

```python
rounds = input("How many turns will this game last?: ")

for i in range(0,int(rounds)):
    A_turn()
    B_turn()
    print_grid()
    print("A Score: " + str(A_points[0]))
    print("B Score: " + str(B_points[0]))
    
if A_points[0] == B_points[0]:
    print("It's a tie!")
elif A_points[0] > B_points[0]:
    print("Player A wins!")
else:
    print("Player B wins!")
```
